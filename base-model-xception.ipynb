## Data Augmentation

```python
"""
This code defines data generators for training, testing, and validation.
It uses the ImageDataGenerator class from Keras to perform data augmentation and preprocessing.

train_datagen_aug: Augmentation datagenerator for training data.
        - shear_range: Range for random shear augmentation.
        - fill_mode: Strategy for filling in newly created pixels during augmentation.
        - rescale: Rescaling factor for pixel values.
        - horizontal_flip: Boolean indicating whether to randomly flip images horizontally.

train_generator_aug: Generator for training data.
        - train_path: Path to the directory containing training images.
        - target_size: Tuple specifying the target size of the images.
        - batch_size: Number of samples per batch.
        - class_mode: Type of label array to be returned.

test_datagen: Datagenerator for testing data.
        - rescale: Rescaling factor for pixel values.

test_generator: Generator for testing data.
        - test_path: Path to the directory containing testing images.
        - target_size: Tuple specifying the target size of the images.
        - batch_size: Number of samples per batch.
        - shuffle: Boolean indicating whether to shuffle the data.
        - class_mode: Type of label array to be returned.

val_generator: Generator for validation data.
        - validation_path: Path to the directory containing validation images.
        - target_size: Tuple specifying the target size of the images.
        - batch_size: Number of samples per batch.
        - class_mode: Type of label array to be returned.
"""
## Augmentation datagenerator
train_datagen_aug = ImageDataGenerator(
        shear_range=0.2,
        fill_mode='nearest',
        rescale=1./255,
        horizontal_flip=True)

## train generator
train_generator_aug = train_datagen_aug.flow_from_directory(
        train_path,
        target_size=(224, 224),
        batch_size=25,
        class_mode='categorical')

## test generator, required for prediction
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
        test_path,
        target_size=(224, 224),
        batch_size=1,
        shuffle = False,
        class_mode='categorical')

## validation generator
val_generator = test_datagen.flow_from_directory(
        validation_path,
        target_size=(224, 224),
        batch_size=10,
        class_mode='categorical')
```

```python
"""
This code calculates the steps per epoch for training and validation data, as well as the number of samples in the test data.
It also assigns the true classes and target names for the test data.
"""

## set steps per epoch for train
train_filenames = train_generator_aug.filenames
steps_train = len(train_filenames)/train_generator_aug.batch_size

## set steps per epoch for validation
val_filenames = val_generator.filenames
steps_val = len(val_filenames)/val_generator_aug.batch_size

## test
filenames_test = test_generator.filenames
nb_samples_test= len(filenames_test)
true_classes = test_generator.classes
target_names = classes
```

# Xception

```python
model = xception.Xception(weights='imagenet', include_top=False, input_shape=(224, 224, 3), classes=3, pooling='avg')


#Get the summary of the model.



model = xception.Xception(weights='imagenet', include_top=False, input_shape=(224, 224, 3),classes=3,pooling='avg')
print (model.summary())
```

```python
#Saves a visual representation of the model architecture to a file.
   
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
count = 0
for layer in model.layers:
    count += 1
count
"""
This code counts the number of layers in the model.
Returns:
    count (int): The total number of layers in the model.
"""
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
x=model.output
x=tf.keras.layers.Dropout(.2)(x)
output=Dense(3,activation='softmax')(x)
base_xception=Model(model.input,output)
"""
This code snippet creates a base model using the Xception architecture for grocery classification.
The model takes the output of a pre-trained model, applies a dropout layer with a dropout rate of 0.2,
and then adds a dense layer with 3 units and softmax activation function for multi-class classification.
The resulting model is stored in the variable 'base_xception'.
"""
```

```python
"""
Compiles the base_xception model with categorical crossentropy loss, RMSprop optimizer, and accuracy metric.
Defines early stopping and model checkpoint callbacks.
Fits the model using a generator for training and validation data.
Returns the fit_generator object.
"""

base_xception.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(learning_rate=0.0001),
              metrics=['acc'])

# Callbacks
es = EarlyStopping(monitor='val_accuracy', min_delta=0.01, patience=5, verbose=1)
mc = ModelCheckpoint("./base_best.h5", monitor='val_accuracy', verbose=1, save_best_only=True)
cb = [es, mc]

fit_generator = base_xception.fit(train_generator_aug,
                                   steps_per_epoch=steps_train,
                                   epochs=50,
                                   validation_data=val_generator,
                                   validation_steps=steps_val,
                                   verbose=1,
                                   callbacks=cb)

fit_generator
```

```python
"""
Plots the training and validation accuracy, as well as the training and validation loss, over the epochs.

Parameters:
- fit_generator: The history object returned by the fit_generator method.

"""

acc = fit_generator.history['acc']
val_acc = fit_generator.history['val_acc']
loss = fit_generator.history['loss']
val_loss = fit_generator.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'b', label='Training acc')
plt.plot(epochs, val_acc,'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

```python
base_xception.save('groceries_base_model.h5')
```

```python
# Convert the model.
converter = tf.lite.TFLiteConverter.from_keras_model(base_xception)
tflite_model = converter.convert()

# Save the model.
with open('base_model_xception.tflite', 'wb') as f:
    f.write(tflite_model)
```

## Predict on test data

```python
base_pred= base_xception.predict(test_generator, steps = nb_samples_test, verbose=2)
base_pred= base_pred.argmax(axis=1)
```

```python
cm = confusion_matrix(base_pred, true_classes)
f, ax= plt.subplots(1,1,figsize=(5,3))
sns.heatmap(cm, annot=True, fmt='g', ax=ax)

# labels, title and ticks
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels') 
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(classes) 
ax.yaxis.set_ticklabels(classes)
```

```python
print(classification_report(true_classes, base_pred, target_names=target_names))
```
